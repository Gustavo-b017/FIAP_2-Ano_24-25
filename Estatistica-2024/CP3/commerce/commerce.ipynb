```python
import csv
import json

# Caminho para o arquivo CSV
csv_file_path = './E-commerce.csv'
# Caminho para o arquivo JSON de saída
json_file_path = './E-commerce.json'

data = []

# Ler o arquivo CSV
with open(csv_file_path, mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        data.append(row)

# Converter os dados para JSON e salvar em um arquivo
with open(json_file_path, mode='w', encoding='utf-8') as json_file:
    json.dump(data, json_file, indent=4, ensure_ascii=False)

print(f"Arquivo JSON salvo em: {json_file_path}")
```

# Expansão dos dados de histórico de compras.

```python
import pandas as pd
import json

with open('./E-commerce.json', 'r', encoding='utf-8') as json_file:
    dados_json = json.load(json_file)

dados = pd.DataFrame(dados_json)

dados_expandidos = []

for index, row in dados.iterrows():
    try:
        compras = json.loads(row['Purchase History'])
    except json.JSONDecodeError:
        continue

    i = 0
    while i < len(compras):
        compra = compras[i]
        
        dados_expandidos.append({
            'ID': row['Customer ID'],
            'idade': row['Age'],
            'genero': row['Gender'],
            'cidade': row['Location'],
            'renda anual': row['Annual Income'],

            'data da compra': compra.get('Purchase Date', compra.get('Purchase Dat')),
            'categoria do produto': compra.get('Product Category', compra.get('Product Category')),
            'preco': compra.get('Price', compra.get('preco')),

            'avaliacao': row['Product Reviews'],
            'tempo ativo': row['Time on Site']
        })
        
        i += 1

dados_expandidos_df = pd.DataFrame(dados_expandidos)

dados_expandidos_df
```

# Análise descritiva 

Selecionando colunas específicas.

```python
df_filtrado = dados_expandidos_df[['idade','genero', 'data da compra','categoria do produto','preco']]
df_filtrado
```

# Medidas descritivas de preço

```python
# Definindo a função para calcular a amplitude
def amplitude(x):
    return (x.max() - x.min())

# Definindo a função para calcular o coeficiente de variação
def coeficiente_variacao(x):
    return 100*(x.std() / x.mean())

def desvio_medio(x):
  return abs(x-x.mean()).mean()

def q1(x):
  return x.quantile(0.25)

def q3(x):
  return x.quantile(0.75)

# Usando aggregate para calcular as estatísticas
result = df_filtrado.aggregate({
    'preco': ['count', 'mean','std','min',q1,'median',q3,'max', amplitude, coeficiente_variacao,desvio_medio],
})

print(result)

result.round(0).transpose()
```
