```python
import pandas as pd
import csv
from io import StringIO

# Lista com os caminhos dos 4 arquivos
arquivos = [
    'Apr-01-2024_Jun-30-2024_1851881538941557.csv',
    'Jul-01-2024_Sep-30-2024_1167474658245288.csv',
    'Oct-01-2024_Dec-30-2024_9285544888165711.csv',
    'Jan-01-2025_Mar-24-2025_2436360143377784.csv'
]

def ler_csv_corrigido(caminho):
    with open(caminho, 'r', encoding='utf-8') as f:
        conteudo = f.read()
    reader = csv.reader(StringIO(conteudo), delimiter=',', quotechar='"')
    linhas = list(reader)
    df = pd.DataFrame(linhas[1:], columns=linhas[0])
    return df

# L√™ e concatena todos os arquivos
dfs_corrigidos = [ler_csv_corrigido(arq) for arq in arquivos]
df_final = pd.concat(dfs_corrigidos, ignore_index=True)

# Exibe as 5 primeiras linhas
display(df_final.head())

# Salva o CSV completo
df_final.to_csv('instagram_completo_formatado.csv', index=False)
```

```python
# Salva o Excel completo
df_final.to_excel('instagram_completo_formatado.xlsx', index=False)
```
