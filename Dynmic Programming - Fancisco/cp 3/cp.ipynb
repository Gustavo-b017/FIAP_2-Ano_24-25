A busca binária é um método eficiente para localizar um valor em uma lista ordenada, dividindo a lista em metades a cada iteração. No código apresentado, o índice do meio é calculado utilizando a divisão inteira entre os índices de início e fim, o que identifica a posição central do intervalo atual da lista. Em cada iteração, o algoritmo compara o elemento na posição do meio com o valor procurado, descartando a metade que não contém o alvo, o que torna o processo muito mais rápido do que uma busca linear.

O funcionamento do cálculo do meio é simples: ele pega a soma dos índices de início e fim, divide por dois e descarta a parte decimal, definindo a posição central do intervalo. Se o valor do meio for menor que o alvo, o algoritmo ajusta o início para continuar a busca na metade direita; se for maior, ajusta o fim para focar na metade esquerda. Esse processo se repete até que o valor seja encontrado ou que o intervalo seja esgotado, demonstrando a eficiência da divisão repetida da lista.

```python
# Lista padrão com 10 carros (pode ser modelos ou placas)
preguica = ["Fusca", "Chevette", "Civic", "Corolla", "Gol", "Uno", "HB20", "Renegade", "Onix", "Siena"]

# Pergunta ao usuário se deseja usar a lista padrão ou inserir os carros manualmente
print("Você deseja usar a lista padrão ou inserir manualmente os carros?")
print("Digite 1 para usar a lista padrão ou 2 para inserir manualmente:")
opcao = input("Sua opção: ")

# Inicializa a pilha que representará o estacionamento
estacionamento = []

if opcao == "1":
    # Se o usuário escolher a lista padrão, copiamos a variável 'preguica'
    estacionamento = preguica.copy()  # Usamos copy() para não alterar a lista original
    print("\nVocê escolheu usar a lista padrão.")
else:
    # Se o usuário preferir inserir manualmente, solicitamos a entrada de 10 carros
    print("\nVocê escolheu inserir manualmente os carros.")
    for i in range(10):
        carro = input(f"Digite o carro {i+1} de 10: ")
        estacionamento.append(carro)

# Exibe a ordem de entrada dos carros
print("\nOrdem de entrada dos carros:")
for carro in estacionamento:
    print(carro)

# Exibe a sequência de saída dos carros usando a estratégia LIFO (Last In, First Out)
print("\nSequência de saída dos carros (LIFO):")
iteracao = 1  # Contador de iterações para exibir o passo a passo
while estacionamento:
    carro_saiu = estacionamento.pop()  # Remove o último carro que entrou
    print(f"Saida do: {carro_saiu}; numero do carro: {iteracao}")
    iteracao += 1
```

Esse código simula um estacionamento usando a estratégia LIFO, onde o último carro a entrar é o primeiro a sair. Inicialmente, o programa pergunta se o usuário deseja utilizar uma lista padrão com 10 carros (armazenada na variável preguica) ou inserir os carros manualmente. Conforme a escolha, a lista de carros é copiada para a variável que representa o estacionamento ou é preenchida com as entradas do usuário. Em seguida, o código exibe a ordem de entrada dos carros, mostrando cada veículo conforme foi adicionado à lista.

Na etapa de saída, o algoritmo utiliza um loop que remove os carros da lista com o método pop(), seguindo a estratégia LIFO. Em cada iteração, o código imprime o nome do carro que saiu junto com um número que representa a ordem de saída (através de um contador). Essa abordagem permite identificar de forma simples qual foi a posição original do carro na sequência de entrada, demonstrando de maneira clara como funciona a lógica de remoção em uma pilha.

```python
# Função para executar o Exercício 1: Busca Binária
def busca_binaria_exercicio():
    print("\nExecutando a Busca Binária (Exercício 1)")
    # Lista fixa e ordenada para a busca e o valor que queremos achar
    lista = [3, 5, 7, 8, 11, 14, 15]
    alvo = 11
    inicio = 0       # Início da área que vamos pesquisar
    fim = len(lista) - 1   # Fim da área que vamos pesquisar
    iteracao = 1     # Contador para mostrar cada passo da busca

    # Enquanto ainda tivermos elementos na área de busca...
    while inicio <= fim:
        # Calcula o índice central usando divisão inteira (descarta as casas decimais)
        meio = (inicio + fim) // 2
        print(f"Iteração {iteracao}: início = {inicio}, fim = {fim}, meio = {meio}, valor = {lista[meio]}")
        
        # Se o valor do meio for igual ao alvo, encontramos o que procurávamos!
        if lista[meio] == alvo:
            print(f"Legal! Alvo {alvo} encontrado no índice {meio}!")
            return
        # Se o valor do meio for menor que o alvo, procuramos na metade direita
        elif lista[meio] < alvo:
            inicio = meio + 1
        # Caso contrário, procuramos na metade esquerda
        else:
            fim = meio - 1
        
        iteracao += 1

    # Se sair do loop, significa que o alvo não foi encontrado
    print(f"O alvo {alvo} não foi encontrado na lista.")

# Função para executar o Exercício 2: Atendimento de Clientes (FIFO)
def atender_clientes_fifo():
    print("\nExecutando o Atendimento de Clientes (FIFO - Exercício 2)")
    opcao = input("Você quer usar a lista padrão de clientes? (s/n): ")
    fila = []
    
    if opcao == 's' or opcao == 'S' or opcao == 'sim':
        # Lista padrão de clientes
        fila = ["Cliente A", "Cliente B", "Cliente C", "Cliente D", "Cliente E"]
        print("Blz, vamos usar a lista padrão de clientes!")
    else:
        quantidade = int(input("Quantos clientes você quer inserir? "))
        for i in range(quantidade):
            cliente = input(f"Digite o nome do cliente {i+1}: ")
            fila.append(cliente)
    
    # Mostra a ordem de chegada dos clientes
    print("\nOrdem de chegada dos clientes:")
    for cliente in fila:
        print(cliente)
    
    # Atende os clientes na ordem em que chegaram (FIFO)
    print("\nAtendendo os clientes (FIFO):")
    atendimento = 1
    while fila:
        cliente_atendido = fila.pop(0)  # Remove o primeiro da fila
        print(f"Atendimento {atendimento}: {cliente_atendido}")
        atendimento += 1

# Função para executar o Exercício 3: Retirada de Carros (LIFO)
def retirar_carros_lifo():
    print("\nExecutando a Retirada de Carros (LIFO - Exercício 3)")
    # Lista padrão de carros (pode ser modelo ou placa)
    preguica = ["Fusca", "Chevette", "Civic", "Corolla", "Gol", "Uno", "HB20", "Renegade", "Onix", "Siena"]
    
    print("Você quer usar a lista padrão ou inserir os carros manualmente?")
    print("Digite 1 para usar a lista padrão ou 2 para inserir manualmente:")
    opcao = input("Sua opção: ")
    
    # A pilha que simula o estacionamento (guarda tuplas com número e nome do carro)
    estacionamento = []
    
    if opcao == "1":
        # Se usar a lista padrão, associa cada carro com seu número original
        estacionamento = list(enumerate(preguica, start=1))
        print("\nShow! Vamos usar a lista padrão.")
    else:
        print("\nBeleza, insira os carros manualmente.")
        for i in range(10):
            carro = input(f"Digite o carro {i+1} de 10: ")
            estacionamento.append((i+1, carro))
    
    # Mostra a ordem de entrada dos carros no estacionamento
    print("\nOrdem de entrada dos carros:")
    for num, carro in estacionamento:
        print(f"Carro {num}: {carro}")
    
    # Retira os carros seguindo a lógica LIFO (o último a entrar sai primeiro)
    print("\nAgora, vamos retirar os carros (LIFO):")
    while estacionamento:
        num_original, carro_saiu = estacionamento.pop()  # Remove o último carro adicionado
        print(f"Saída: {carro_saiu} (Número do carro: {num_original})")

# Função principal que permite ao usuário escolher qual estrutura utilizar
def main():
    print("Escolha a estrutura que você quer usar:")
    print("1 - Ordenar a lista do Exercício 1 (Busca Binária)")
    print("2 - Atender os clientes do Exercício 2 (teorico) (FIFO)")
    print("3 - Retirar os carros do Exercício 3 (LIFO)")
    
    opcao = input("Sua opção: ")
    
    if opcao == "1":
        busca_binaria_exercicio()
    elif opcao == "2":
        atender_clientes_fifo()
    elif opcao == "3":
        retirar_carros_lifo()
    else:
        print("Opção inválida. Tente novamente.")

# Executa o programa principal
if __name__ == "__main__":
    main()

# OBS: nao existe o exercicio 2, entao fiz parecido com o 3, mas mudando a aplicação espero q n se importe (*^▽^*)
```
